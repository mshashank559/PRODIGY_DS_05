```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
file_path = os.path.join('..', 'Cleaned data', 'processed_data.csv')  # go up one level

if os.path.exists(file_path):
    df = pd.read_csv(file_path)
    print("Data loaded successfully. Shape:", df.shape)
else:
    print(f"File not found: {file_path}")
```

```python
os.makedirs('visuals', exist_ok=True)
```

```python
# 1. Accidents by Day of the Week
plt.figure(figsize=(10, 6))
sns.countplot(x='DayOfWeek', data=df, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], palette='cubehelix')
plt.title('Accidents by Day of the Week')
plt.ylabel('Number of Accidents')
plt.xlabel('Day')
plt.tight_layout()
plt.savefig('visuals/accidents_by_day.png')
plt.show()
```

```python
# 2. Severity by Hour of Day
plt.figure(figsize=(12, 6))
sns.boxplot(x='Hour', y='Severity', data=df, palette='Set2')
plt.title('Accident Severity by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Severity')
plt.tight_layout()
plt.savefig('visuals/severity_by_hour.png')
plt.show()
```

```python
# 3. Top 10 States with Most Accidents
if 'State' in df.columns:
    top_states = df['State'].value_counts().nlargest(10)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=top_states.index, y=top_states.values, palette='flare')
    plt.title('Top 10 States with Most Accidents')
    plt.ylabel('Number of Accidents')
    plt.xlabel('State')
    plt.tight_layout()
    plt.savefig('visuals/top_states.png')
    plt.show()
```
